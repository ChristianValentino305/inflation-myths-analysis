```python
sentiment= pd.read_csv("/content/drive/MyDrive/Info Visualization/sentiment.csv")
sentiment
```

```python
melted_sentiment = pd.melt(
    sentiment,
    id_vars=["Year"],
    value_vars=["UMCSENT_PC1","CPI"],
    var_name="type",
    value_name="value"
)
melted_sentiment
```

```python
 melted_sentiment['type'] = melted_sentiment['type'].replace({
    "UMCSENT_PC1": "Consumer Sentiment",
    "CPI": "Inflation"
})


color_scale = alt.Scale(
    domain=['Consumer Sentiment', 'Inflation'],
    range=['#007A7A', '#CB0000']
)

sentiment_lines = alt.Chart(melted_sentiment).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='')),
    y=alt.Y('value:Q', axis=alt.Axis(title='% Change', format='.1f')),
    color=alt.Color('type:N', scale=color_scale, legend=alt.Legend(
        title='',
        values=['Consumer Sentiment', 'Inflation'], labelFontSize=11, symbolSize=170))
)

zeroY = alt.Chart(pd.DataFrame({'y': [0]})).mark_rule(color='black', strokeDash=[7, 3]).encode(
    y='y:Q'
)

sentimentchart=sentiment_lines+zeroY

sentimentchart=sentimentchart.properties(
    width=700,
    height=300,
    title='Changes in Consumer Sentiment and Inflation'
)
sentimentchart
```

```python
html = sentimentchart.to_html()

with open('/content/sentiment_chart.html', 'w') as f:
    f.write(html)
```

```python
melted_expectations['type'] = melted_expectations['type'].replace({
    "Consumer_5year": "Consumers",
    "Econ_5year": "Economists",
    "BreakEven_5year": "Breakeven Rate",
    "CPI_5year": "Inflation"
})

legend_select = alt.selection_point(fields=['type'], bind='legend')

color_scale = alt.Scale(
    domain=['Consumers', 'Economists', 'Breakeven Rate', 'Inflation'],
    range=['#007A7A', '#CB5C00', '#00A300', '#CB0000']
)

expectation_lines = alt.Chart(melted_expectations).mark_line().encode(
    x=alt.X('Year5:T', axis=alt.Axis(title='')),
    y=alt.Y('value:Q', axis=alt.Axis(title='Inflation (%)', format='.1f')),
    color=alt.Color('type:N', scale=color_scale, legend=alt.Legend(
        title='(Click to Highlight)',
        values=['Consumers', 'Economists', 'Breakeven Rate', 'Inflation'], labelFontSize=11, symbolSize=160)),
    opacity=alt.condition(legend_select, alt.value(1), alt.value(0.2))
).add_params(
    legend_select
)

inflation_line = alt.Chart(melted_expectations[melted_expectations['type'] == 'Inflation']).mark_line(
    color='#CB0000',
    strokeWidth=2
).encode(
    x='Year5:T',
    y='value:Q'
)

expectationlines=expectation_lines+inflation_line.properties(
    width=600,
    height=300,
    title='5-Year Ahead Inflation Forcasts vs Actual Inflation'
)

expectationlines
```

```python
html = expectationlines.to_html()

with open('/content/expectationlines.html', 'w') as f:
    f.write(html)

from google.colab import files
files.download('/content/expectationlines.html')
```

```python
melted_expectations_dif = pd.melt(
    expectations,
    id_vars=["Year5"],
    value_vars=["Consumer_dif5", "Econ_dif5", "Breakeven_dif5"],
    var_name="type",
    value_name="value"
)

melted_expectations_dif['type'] = melted_expectations_dif['type'].replace({
    "Consumer_dif5": "Consumers",
    "Econ_dif5": "Economists",
    "Breakeven_dif5": "Breakeven Rate"
})

color_scale = alt.Scale(
    domain=['Consumers', 'Economists', 'Breakeven Rate'],
    range=['#007A7A', '#CB5C00', '#00A300']
)

density_chart = alt.Chart(melted_expectations_dif).transform_density(
    density='value',
    groupby=['type'],
    as_=['Difference From Actual Inflation', 'Frequency']
).mark_line().encode(
    x=alt.X('Difference From Actual Inflation:Q', axis=alt.Axis(format='.1f')),
    y='Frequency:Q',
    color=alt.Color('type:N', scale=color_scale, legend=alt.Legend(title='(Click to Highlight)', labelFontSize=11, symbolSize=160)),
        opacity=alt.condition(legend_select, alt.value(1), alt.value(0.2))
).add_params(
    legend_select
)

zero = alt.Chart(pd.DataFrame({'x': [0]})).mark_rule(color='#CB0000', strokeDash=[7, 3]).encode(
    x=alt.X('x:Q', axis=alt.Axis(title="Difference from Actual Inflation", format='.1f'))
)

accuracy_chart = (density_chart + zero).properties(
    width=600,
    height=300,
    title='Accuracy of 5-year Ahead Inflation Forecasts'
)
accuracy_chart
```

```python
html = accuracy_chart.to_html()

with open('/content/accuracy_chart.html', 'w') as f:
    f.write(html)

from google.colab import files
files.download('/content/accuracy_chart.html')
```
